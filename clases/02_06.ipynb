Varios plots dentro de un for:

```julia
using Plots
gr()
```

```julia
for i in 1:5
    plot(x -> sin(x + i))  #  función anónima; cf. ↦
end
```

No aparece ningún plot, ya que el `for` no regresa nada.

```julia
p = plot()

for i in 1:5
    scatter!(x -> sin(x + i), 0, 5)  #  función anónima; cf. ↦
    plot!(x -> sin(x + i), 0, 5)  #  función anónima; cf. ↦

end
```

```julia
plot!()
```

```julia
typeof(p)
```
